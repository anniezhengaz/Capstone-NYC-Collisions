```python
#Import necessary functions
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from datetime import datetime as dt

%matplotlib inline
```

```python
#Import the dataset
df = pd.read_csv('data/Motor_Vehicle_Collisions_QuBr_2018-2023.csv')

#Initial look into dataset
df.head()
```

```python
df.info()
```

#### Changing DateTime Index

```python
#Set to datetime index
df.set_index('CRASH DATE TIME', inplace=True)
```

```python
df.head()
```

## EDA

```python
df.info()
```

```python
#Shift 'Collision ID' column to first position
first_column = df.pop('COLLISION_ID')
  
# insert column using insert(position,column_name,first_column) function
df.insert(0, 'COLLISION_ID', first_column)
```

#### Dropping Columns

```python
#drop additional contributing factor vehicle columns
df.drop(['CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5'], axis=1, inplace=True)

#drop additional vehicle type columns
df.drop(['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2', 'VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
#Drop NaN values from columns of contributing factors
df = df.dropna(subset=['CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2'])
```

```python
df.info()
```

```python
df.head()
```

```python
#Check for NaN values in injury/death columns
print(df['NUMBER OF PERSONS KILLED'].isnull().sum())
print(df['NUMBER OF PERSONS INJURED'].isnull().sum())
```

```python
#Drop NaN values from injury/death columns
df = df.dropna(subset=['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED'])

#Sanity check to verify no more NaN values
print(df['NUMBER OF PERSONS KILLED'].isnull().sum())
print(df['NUMBER OF PERSONS INJURED'].isnull().sum())
```

```python
df.info()
```

```python
#Filter out the collisions with Unspecified contributing factor
df = df[df['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']
df = df[df['CONTRIBUTING FACTOR VEHICLE 2'] != 'Unspecified']

df
```

```python
#Drop missing values from Latitude and Longitude columns -- ensure we can map
df = df.dropna(subset=['LATITUDE', 'LONGITUDE'])
df.info()
```

```python
#Check rows with missing zip code
df[df['ZIP CODE'].isna()]
```

```python
#df['ZIP CODE'].replace(('Dwe', 'DW', 'TWE', 'dwe', 'DWE/Anglican church', 'RWE'), ('DWE'), inplace=True)
```

```python
#Sanity check to confirm zipcode was imputed for indexed row
df[df.index.isin(['2018-01-18 17:20:00'])]
```
