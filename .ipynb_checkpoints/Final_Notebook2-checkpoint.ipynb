## Modeling

```python
#Make list for leading causes
leading_causes = ['Moving Violation', 'Driver Inattention']
```

### Stationarity
Before we can begin modeling, we must check the stationarity of the dataset. A stationary dataset is required for Time Series models to perform well and accurately. The Dickey-Fuller statistical test will be utilized to verify stationarity.

```python
#For loop to test stationarity
for cause in leading_causes:
    #Test for stationarity with the adfuller test
    ts_cause = ts_pivt[cause]
    p_val = adfuller(ts_cause)[1]
    print(f"The p-value associated with the Dickey-Fuller statistical test for {cause} is {p_val},")
    if p_val < 0.05:
        print(" therefore we can safely assume that the data is stationary.")
    else:
        print(" therefore we cannot reject the null hypothesis that the data is not stationary.")
        
    #Plot the dataset
    fig, ax = plt.subplots(figsize=(12, 5))
    ts_cause.plot(ax=ax)
    ax.set_title(f"Time Series Plot for {cause}")
    ax.set_xlabel("Crash Date")
    ax.set_ylabel("Number of Collisions")
    plt.show()
```

### Differencing
There are numerous methods to making a dataset stationary, however the method we will be using for this dataset is differencing.

```python
#For loop for differencing
for cause in leading_causes:
    #Difference the data and drop the NaN values
    diff = ts_cause.diff().dropna()
    
    print("\nResults after differencing:")
    # Augmented Dickey-Fuller test
    p_val_diff = adfuller(diff)[1]
    print(f"The p-value associated with the Dickey-Fuller statistical test for {cause} is {p_val_diff},")
    if p_val_diff < 0.05:
        print("therefore we can safely assume that the differenced data is stationary.")
    else:
        print("therefore we cannot reject the null hypothesis that the differenced data is not stationary.")
    
    #Plot the differenced dataset
    fig, ax = plt.subplots(figsize=(12, 5))
    diff.plot(ax=ax)
    ax.set_title(f"Time Series Plot for {cause}")
    ax.set_xlabel("Crash Date")
    ax.set_ylabel("Number of Collisions")
    plt.show()
```

### Train Test Split

```python
#For loop for train-test split
for cause in leading_causes:
    #Find the index which allows us to split off 20% of the data
    cutoff = round(ts_cause.shape[0]*0.8) #cut off point at 80% of the data

    # Define train and test sets according to the index found above
    train = ts_cause[:cutoff] #before cut off is train set -- 80%

    test = ts_cause[cutoff:] #after cut off is the test set --20%

    fig, ax = plt.subplots(figsize=(12, 5))
    ax.plot(train, label='train')
    ax.plot(test, label='test')
    ax.set_title(f'Train-Test Split for {cause}')
    plt.legend();
```

### Baseline Model


```python
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
for cause in leading_causes:
    # Baseline model
    baseline = train.copy()
    baseline = baseline.to_frame()
    baseline['One Shift'] = baseline.shift(1)
    baseline.plot(figsize=(12, 5))
    plt.title(f'Baseline Model for {cause}')
    
    #A closer look
    baseline[0:30].plot(figsize=(12, 5))
    plt.title(f' Closer look at the Baseline Model for {cause}')
    
    #Calculate the Root Mean Square Error of the baseline model
    # Calculate RMSE of baseline model
    #y_true = test[cause].values
    #y_pred = baseline['One Shift'][cutoff:].values
    #baseline_rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"{cause} Baseline Model RMSE: {baseline_rmse:.2f}")
    #y_true = test.values
    #y_pred = baseline['One Shift'][cutoff:].values
    #mean_absolute_percentage_error(baseline, y_pred)
    
    # calculate RMSE for baseline model
    #y_true = test.values
    #y_pred = baseline['One Shift'][cutoff:].values
    #rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    #print(f"{cause} Baseline Model RMSE: {rmse:.2f}")
```

```python
# Calculate RMSE of baseline model
y_true = test[cause].values
y_pred = baseline['One Shift'][cutoff:].values
baseline_rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f"{cause} Baseline Model RMSE: {baseline_rmse:.2f}")
```
