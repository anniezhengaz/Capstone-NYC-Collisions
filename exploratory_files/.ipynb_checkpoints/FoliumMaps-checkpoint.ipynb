## Folium Maps

```python
df
```

```python
#Create dataset with only Queens collisions
Queens = df[df['BOROUGH'] == 'QUEENS']

#Create dataframe with only relevant coliums to map making
Queens_Collisions = Queens[['LATITUDE', 'LONGITUDE', 'COLLISION_ID']]
Queens_Collisions
```

```python
#Create a blank map
map = folium.Map(location=[Queens_Collisions.LATITUDE.mean(), Queens_Collisions.LONGITUDE.mean()], zoom_start=14, control_scale=True)

#Create index markers for Queens collisions
for index, location_info in Queens_Collisions.iterrows():
    folium.Marker([location_info["LATITUDE"], location_info["LONGITUDE"]], popup=location_info["COLLISION_ID"]).add_to(map)
    
map
```
